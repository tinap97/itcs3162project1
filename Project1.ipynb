```python
import pandas as pd
```

```python
studentperf_df = pd.read_csv("StudentsPerformance.csv")
```

```python
studentperf_df['average scores'] = (studentperf_df['math score'] + studentperf_df['reading score'] + studentperf_df['writing score']) / 3
```

```python
studentperf_df.head()
```

```python
studentperf_df.drop('race/ethnicity', axis = 1, inplace=True)
```

```python
studentperf_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
studentperf_df.plot(x='writing score', y='parental level of education', kind='scatter')
plt.title("Writing Score Based on Parent Level of Education")
f = plt.figure()
f.set_figwidth(200)
f.set_figheight(50)
```

```python
studentperf_df.plot(x='average scores', y='parental level of education', kind='scatter')
plt.title("Average Test Scores Based on Parent Level of Education")
f = plt.figure()
f.set_figwidth(200)
f.set_figheight(5)
```

```python
studentperf_df.plot(x='math score', y='parental level of education', kind='scatter')
plt.title("Math Score Based on Parent Level of Education")
```

```python
studentperf_df.plot(x='reading score', y='parental level of education', kind='scatter')
plt.title("Reading Score Based on Parent Level of Education")
```

```python
studentperf_df.plot(x='average scores', y='lunch', kind='scatter')
plt.title("Average Scores Based on lunch")
```
